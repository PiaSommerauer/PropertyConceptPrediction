```python
import math
import csv
from collections import defaultdict
pd.set_option('display.max_rows', None)
```

```python
# instace-based comparison

prop = 'wings'
model = 'bert-base-uncased'

concept_dict = defaultdict(dict)

targets = ['concept', 'property']

for target in targets:
    path = f'../results/{target}/{model}/probabilities/{prop}.csv'
    print(path)
    with open(path) as infile:
        data = list(csv.DictReader(infile))
    for d in data:
        concept = d['concept']
        label = d['label']
        p = float(d['prob'])
        p_prior = float(d['prob_control'])
        concept_dict[concept][('label', 'label')] = label
        concept_dict[concept][(target, 'prob_norm')] = round(math.log(p/p_prior), 4)
        concept_dict[concept][(target, 'prob')] = round(p, 4)
```

```python
df = pd.DataFrame(concept_dict).T.sort_values(('concept', 'prob_norm'), ascending = False)

print(df.round(4)[:20].to_latex())
#df
```
